#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui

media= []

for numero in novo_arr_df['media']:
    media.append(numero)


log_media=[]

for log in media:
    resultado=math.log(log)
    log_media.append(resultado)



log_media
```
